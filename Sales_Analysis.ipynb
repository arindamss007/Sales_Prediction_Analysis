## **SALES PREDICTION & ANALYSIS**

NOTE : *THIS ANALYSIS IS BASE ON THE DATA, THAT IS ALREADY PROVIDED. IN THIS ANALYSIS, IT IS QUITE A CHALLANGE TO BUILD A ML MODEL. TO BUILD A ML MODEL, IT'S NEED MORE DATA.*

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def plot_continuous_distribution(data: pd.DataFrame = None, column: str = None, height: int = 8):
  _ = sns.displot(data, x=column, kde=True, height=height, aspect=height/5).set(title=f'Distribution of {column}');

df = pd.read_csv('/content/Advertising.csv')

df = df.drop(columns = ['Unnamed: 0'])

df.head()
```

**DATA DESCRIPTIONS**

```python
df.describe()
```

**COLUMN RELATIONSHIP**

```python
df.corr()
```

**PURCHASE COUNT VIA TELEVISION(TV) ADVERTISEMENT**

```python
plot_continuous_distribution(df, 'TV',6)
```

**PURCHASE COUNT VIA RADIO ADVERTISEMENT**

```python
plot_continuous_distribution(df, 'Radio',5)
```

**PURCHASE COUNT VIA NEWSPAPER ADVERTISEMENT**

```python
plot_continuous_distribution(df, 'Newspaper',6)
```

**PURCHASE COUNT VIA REGULAR SALES**

```python
plot_continuous_distribution(df, 'Sales',5)
```

**AVARAGE PURCHASE VISUALIZATION**

```python
avgdata = {
    "INDEX" : ['TV', 'Radio', 'Newspaper', 'Sales'],
    "Average" : [147.0425, 23.264000000000006, 30.553999999999995, 14.0225]
    }

avg = pd.DataFrame(avgdata)

plt.barh(avg['INDEX'], avg['Average'])
plt.title("Avarage Purchase Graph")
plt.xlabel("Quantity")
plt.ylabel("Factors")
plt.show()
```

**MINIMUM PURCHASE VISUALIZATION**

```python
mindata = {
    "INDEX" : ['TV', 'Radio', 'Newspaper', 'Sales'],
    "MIN" : [0.7, 0, 0.3, 1.6]
    }

min = pd.DataFrame(mindata)

plt.barh(min['INDEX'], min['MIN'])
plt.title("Minimum Purchase Graph")
plt.xlabel("Quantity")
plt.ylabel("Factors")
plt.show()
```

**MAXIMUM PURCHASE VISUALIZATION**

```python
maxdata = {
    "INDEX" : ['TV', 'Radio', 'Newspaper', 'Sales'],
    "MAX" : [296.4, 49.6, 114, 27]
    }

max = pd.DataFrame(maxdata)

plt.barh(max['INDEX'], max['MAX'])
plt.title("Maximum Purchase Graph")
plt.xlabel("Quantity")
plt.ylabel("Factors")
plt.show()
```
